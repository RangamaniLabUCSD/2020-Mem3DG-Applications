## Geometry and basic setup

```python
import sympy as sp
import numpy as np
import pymem3dg as dg 

r = sp.symbols('r')
sp.init_printing(use_unicode=True)
# z = 1 * sp.sqrt(1-r**2) 

R = 1
c0 = 0.2072
c1 = 2.0026
c2 = -1.1228
z = 0.5 * R * sp.sqrt(1 - r**2) * (c0 + c1 * r**2 + c2 * r**4)
sp.plotting.plot(z, xlim=[0,1])
```

## Differentiation

```python
zPrime = sp.diff(z,r)
zPrimePrime = sp.diff(zPrime,r)
dsdr = sp.sqrt(1 + zPrime**2)
drds = 1 / dsdr
```

## Curvatures

```python
kappaNu = drds * zPrime / r
kappaT = drds**3 * zPrimePrime
meanCurvature = -0.5 * (kappaNu + kappaT)
gaussianCurvature = kappaNu * kappaT

# plotting 
meanCurvature
# sp.plotting.plot(meanCurvature, xlim=[0,1], ylim=[-2,3])
# plotting.plot(kappaNu, xlim=[0,R])
# plotting.plot(kappaT, xlim=[0,R])
```
